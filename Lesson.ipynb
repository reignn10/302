# Import Libraries

```python
import pandas as pd #Data Manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modeling
```

# Generating Random Data

```python
num_points = 50

X = np.random.randint (0,100,num_points)

Y = 2 * X + np.random.normal (0,10,num_points)
```

# Creating DataFrame and Saving to CSV

```python
data  = pd.DataFrame ({'X': X, 'Y': Y})

data.to_csv('data.csv', index=False)

print(data.head())
```

Visualizing the Data

```python
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable (X)')
plt.ylabel('Independent Variable (Y)')
plt.title('Magulong Plot')
plt.show()
```

Preparing Data for Monitoring

```python
X = data[['X']] #Indipendent Variable
Y = data[['Y']] #Dependent Variables
```

Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(X,Y)
```

Getting Model Coefficient

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope',slope)
print('Intercept', intercept)
```

Visualizing the Linear Regression Line

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color= 'red')
plt.xlabel('Independet Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Linear Regression')
plt.show()
```

Making Predictions

```python
new_X = [[np.random.randint(100,200)]]

predict_Y = model.predict(new_X)

print('Value of X:  ',new_X)
print('Predicted Value of Y  ', predict_Y)
```
